```python
#reading all the files 
# importing the required modules
import glob
import pandas as pd
  
# specifying the path to csv files
path = os.getcwd()
  
# csv files in the path
files = glob.glob(path + "/*.csv")
  
# defining an empty list to store 
# content
data_frame = pd.DataFrame()
content = []
  
# checking all the csv files in the 
# specified path
for filename in files:
    
    # reading content of csv file
    # content.append(filename)
    df = pd.read_csv(filename, index_col=None)
    content.append(df)
  
# print(content[1])
# converting content to data frame
data_frame = pd.concat(content, sort=False)
data_frame
```

```python
# saving the dataframe
data_frame.to_csv('final_game_of_throns1.csv',index=False)
# saving the dataframe to specified path
#df.to_csv(r'C:\Users\Admin\Desktop\file3.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("battles (1).csv")
pd.pandas.set_option('display.max_columns', None) 
df.head()
```

```python
df['battle_type'].value_counts()#shows the value count for all individual battle types..so there are 4 types of battles
```

```python
for index,value in enumerate(df['battle_type'].value_counts().values):
    print(index,value)
```

```python
fig,ax1=plt.subplots(1,1,figsize=(10,10))
sns.barplot(x=df['battle_type'].value_counts().index,y=df['battle_type'].value_counts().values,color="orangered",label="Battle")
for index,value in enumerate(df['battle_type'].value_counts().values):
    ax1.annotate(value,xy=(index,value),ha="center",va="center",fontsize=18)
plt.legend(fontsize=26)
plt.xlabel("Battle_type",fontsize=20)
plt.ylabel("Battle_count",fontsize=20)
plt.title("various Battle type",fontsize=30)
plt.show()
```
